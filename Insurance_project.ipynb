Project on Insurance prediction,insurance amount going to get for new customers, from the health insurance company.
Step No;1 Gathering the data

```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("/content/insurance.csv")
```

```python
data.head()
```

Step 2: Data cleaning

```python
#converting categorical variables into numerical format-One Hot Encoding
data['region']=data["region"].apply({'southwest':1,'southeast':2,'northeast':3,'northwest':4}.get)
data['sex']=data["sex"].apply({'female':1,'male':0}.get)
data['smoker']=data["smoker"].apply({'yes':1,'no':0}.get)
```

```python
data.head()
```

Step 3: Deviding the Data into dependent and independent variables

```python
x=data[['age','sex','bmi','children','smoker','region']] #independent variables
y=data['expenses']#Dependent variables
```

Step:4 Splitting the dataset into training and testing set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.80)
```

Creating our SMLM using Linear Regression algorithm

```python
from sklearn.linear_model import LinearRegression
regression=LinearRegression() # we are activating our algorithm
```

```python
regression.fit(x_train,y_train) #.fit used for training
```

```python
y_pred=regression.predict(x_test)
```

```python
print(y_pred)
```

```python
#accuracy of SMLM
regression.score(x,y)
```
